```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris = load_iris()
mean = np.mean(iris.data)
sigma = np.std(iris.data)
print('The mean of the dataset is',mean)
print('The standard deviation of the dataset is',sigma)
def normalize(iris):
    return (iris-mean)/sigma
z = normalize(iris.data)
z_mean = np.mean(z)
print('Standardized centered and scaled z_mean is',z_mean)
z_std = np.std(z)
print('The standard deviation of the z is',z_std)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def normalize(x, mu, sigma):
    return np.exp(-(x - mu)**2 / (2 * sigma**2)) / (sigma * np.sqrt(2 * np.pi))

x = np.linspace(-10, 10, 100)
y1 = normalize(x, 0, 0.5)
y2 = normalize(x, 0, 1)
y3 = normalize(x, 0, 2)

plt.plot(x, y1, label='sigma = 0.5')
plt.plot(x, y2, label='sigma = 1')
plt.plot(x, y3, label='sigma = 2')
plt.xlabel('x')
plt.ylabel('Normalized y')
plt.title('Normalized function')
plt.legend()
plt.show()
```
